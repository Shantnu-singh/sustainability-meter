```python
# this is to calculate score for age grp based on the option they chhose ( most prob won't have to change this )
def calculate_scores(assigned_questions, assigned_color):
    total_score = 0

    for i in range(len(assigned_questions)):
        options = [option for option, score in scoring_info[assigned_color[i]]]

        print(f"Question {i + 1}: {assigned_questions[i]}")
        print("Options:")
        for idx, option in enumerate(options):
            print(f"{idx + 1}. {option}")

        chosen_option_index = int(input("Choose an option (enter the option number): ")) - 1

        while chosen_option_index < 0 or chosen_option_index >= len(options):
            print("Invalid option number. Please choose again.")
            chosen_option_index = int(input("Choose an option (enter the option number): ")) - 1

        score = scoring_info[assigned_color[i]][chosen_option_index][1]
        total_score += score
        print("\n")

    return total_score
```

```python
age = int(input("Enter the age: "))
assigned_questions, assigned_color = assign_questions(age)
total_score = calculate_scores(assigned_questions, assigned_color)
print(f"Total Score: {total_score}")
```
